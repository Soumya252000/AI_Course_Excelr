```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

year = [1950,1960,1970,1980,1990,2000]
gdp = [300,540,600,700,200,300]
```

```python
plt.figure(figsize=(6,3))
plt.xlabel("year")
plt.ylabel("GDP", color='blue')
plt.title("GDP over the years",fontsize = 18)
plt.plot(year,gdp)
plt.show()
```

```python
google = pd.read_csv('datasets/Google.csv',index_col = 'Date', parse_dates=True, usecols = ['Date','Close'])
tesla = pd.read_csv('datasets/Tesla.csv',index_col = 'Date', parse_dates=True, usecols = ['Date','Close'])
tesla
```

```python
plt.title("Google vs Tesla")
plt.xlabel('year')
plt.ylabel('Close')
plt.plot(google,label='google')
plt.plot(tesla,label='tesla')
plt.legend()
```
