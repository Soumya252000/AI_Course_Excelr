#### Linear Regression

```python
### Clean the data
rdata = pd.read_csv(r'C:\Users\sbadiger\AI_Practice_Excelr\datasets\insurance.csv')
rdata.head()
```

```python
rdata.info()
```

```python
### for object data standardise and for other data standardise the data.
from sklearn.preprocessing import LabelEncoder, StandardScaler
enc = LabelEncoder()
scaler = StandardScaler()
for column in rdata.columns:
    if rdata[column].dtype == 'object':
        rdata[column] = enc.fit_transform(rdata[column])
    else:
        rdata[column] = scaler.fit_transform(rdata[[column]])
rdata.head()
```

```python
### Split the data
X = rdata.drop('charges',axis=1)
Y = rdata['charges']
```

```python
### Build the model and adding layers
### No of neurons is based on no of features - 6neurons and 8 hidden layer
model = Sequential(
   [Input(shape=(X.shape[1],)),
    Dense(8,activation='relu'),
    Dense(8,activation='relu'),
    Dense(1,activation='linear')]
)
```
